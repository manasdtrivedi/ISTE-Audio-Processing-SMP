### ${\frac{1}{3^n}}u[n]$

```python
#write code here
import numpy as np
import matplotlib.pyplot as pyplot
x = np.arange(-20,21,1)
print(x)
y=[]
for i in x:
    if i<0:
        y.append(0)
    else:
        y.append(np.power((1/3),i))       
        
pyplot.stem(x,y)
pyplot.show()
```

### $3 \cos \left[ \frac{\pi n}{4} + \frac{\pi}{3} \right]$

```python
#write code here
import numpy as np
import matplotlib.pyplot as pyplot
x = np.arange(-20,21,1)

y=np.cos(np.pi*x/4+np.pi/3)
      
        
pyplot.stem(x,y)
pyplot.show()
```

### $(-0.75)^n u[n]$

### $\frac{\sin (\pi n/6 ) }{\pi n/6}$

```python
#write code here
import numpy as np
import matplotlib.pyplot as pyplot
x = np.arange(-20,21,1)
y=[]
for i in x:
    if i==0:
        y.append(1)
    else:
        y.append((np.sin(np.pi*i/6))/(np.pi*i/6))

pyplot.stem(x,y)
pyplot.show()
```

### $cos\left( \frac{\pi 5n}{7} \right)+cos\left( \frac{\pi n}{5} \right)$

```python
#write code here
import numpy as np
import matplotlib.pyplot as pyplot
x = np.arange(-20,21,1)

y=np.cos(np.pi*5*x/7)+np.cos(np.pi*x/7)

pyplot.stem(x,y)
pyplot.show()
```

### ${\sin (\pi n^2 / 200 ) }$

```python
import numpy as np
import matplotlib.pyplot as pyplot

##for power use x**n where n is power
##or use np.power
##make sure that at n=0,function value is 1

x = np.arange(-20,21,1)
y=[]
for i in x:
    if i==0:
        y.append(1)
    else:
        y.append(np.sin(np.pi*(i**2)/200))

pyplot.stem(x,y)
pyplot.show()
```

# t=0 to 1,
plot x(t)+x(t+5),where x is unit step function

```python
##code here
import numpy as np
import matplotlib.pyplot as pyplot
t=np.arange(0,1.1,0.1)
x=[]
for i in t:
    if i<-5:
        x.append(0)
    elif i<0:
        x.append(1)
    else:
        x.append(2)
        
pyplot.stem(t,x)
pyplot.show()
```
